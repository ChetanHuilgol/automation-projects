### Rename Files in Bulk

- In this section, we will load all the files from a specific directory and view the files.

```r
old_path = 'data/old_files'
file_type = "*.jpg"
old_files = list.files(old_path, pattern=file_type, full.names=TRUE)
old_files
```

- Next, we will rename the files based on a specific logical value

```r
new_path = 'data/new_files'
standard_name = paste0(new_path, '/wallpaper ')
new_files = paste0(standard_name, 1:length(old_files), ".JPG")
new_files
```

- Finally, we will move all the files from one directory to another.

```r
file.copy(from=old_files, to=new_files)
```
